```python
# 运行 python tools/train.py --config_file configs/omnire.yaml --output_root /mnt/e/Output/ --project cam3 --run_name front data.scene_idx=23

import os
import subprocess

def preprocess_data(config_file, output_root, project, run_name, scene_idx):

    command = [
        "python", "tools/train.py",
        "--config_file", config_file,
        "--output_root", output_root,
        "--project", project,
        "--run_name", run_name,
        f"data.scene_idx={str(scene_idx)}",
    ]
    # 在 .. 目录下运行命令
    subprocess.run(command, check=True, cwd="..")

# Example usage
for id in [23, 114, 327, 621, 703, 172, 552, 788]:
    preprocess_data(config_file="configs/omnire_extended_cam.yaml", output_root="/mnt/e/Output/", project="cam3", run_name=str(id), scene_idx=id)
```
